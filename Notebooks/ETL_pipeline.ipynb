```python
orders_clean = orders.copy()

# Convert date columns to datetime
date_cols = ['order_purchase_timestamp','order_approved_at',
             'order_delivered_carrier_date','order_delivered_customer_date',
             'order_estimated_delivery_date']
for col in date_cols:
    orders_clean[col] = pd.to_datetime(orders_clean[col], errors='coerce')

# Drop rows with missing order_id or customer_id
orders_clean = orders_clean.dropna(subset=['order_id','customer_id'])

print("Orders cleaned shape:", orders_clean.shape)
orders_clean.head()
```

```python
customers_clean = customers.copy()

# Drop duplicates
customers_clean = customers_clean.drop_duplicates()

# Strip strings
for col in ['customer_unique_id','customer_zip_code_prefix']:
    customers_clean[col] = customers_clean[col].astype(str).str.strip()

print("Customers cleaned shape:", customers_clean.shape)
customers_clean.head()
```

```python
order_items_clean = order_items.copy()

# Convert price and freight to numeric
order_items_clean['price'] = pd.to_numeric(order_items_clean['price'], errors='coerce')
order_items_clean['freight_value'] = pd.to_numeric(order_items_clean['freight_value'], errors='coerce')

# Drop rows missing mandatory keys
order_items_clean = order_items_clean.dropna(subset=['order_id','product_id','seller_id'])

print("Order items cleaned shape:", order_items_clean.shape)
order_items_clean.head()
```

```python
products_clean = products.copy()
products_clean = products_clean.drop_duplicates()

# Convert product weight/length etc. to numeric
num_cols = ['product_weight_g','product_length_cm','product_height_cm','product_width_cm']
for col in num_cols:
    products_clean[col] = pd.to_numeric(products_clean[col], errors='coerce')

print("Products cleaned shape:", products_clean.shape)
products_clean.head()
```

```python
payments_clean = payments.copy()

# Convert payment_value to numeric
payments_clean['payment_value'] = pd.to_numeric(payments_clean['payment_value'], errors='coerce')

# Drop rows with null order_id
payments_clean = payments_clean.dropna(subset=['order_id'])

print("Payments cleaned shape:", payments_clean.shape)
payments_clean['payment_type'].value_counts()
```

```python
sellers_clean = sellers.copy()
sellers_clean = sellers_clean.drop_duplicates()

print("Sellers cleaned shape:", sellers_clean.shape)
sellers_clean.head()
```

```python
geolocation_clean = geolocation.copy()

# Drop duplicates on zipcode
geolocation_clean = geolocation_clean.drop_duplicates(subset=['geolocation_zip_code_prefix'])

print("Geolocation cleaned shape:", geolocation_clean.shape)
geolocation_clean.head()
```

```python
customers_clean.to_csv("customers_clean.csv", index=False)
orders_clean.to_csv("orders_clean.csv", index=False)
order_items_clean.to_csv("order_items_clean.csv", index=False)
products_clean.to_csv("products_clean.csv", index=False)
payments_clean.to_csv("payments_clean.csv", index=False)
sellers_clean.to_csv("sellers_clean.csv", index=False)
geolocation_clean.to_csv("geolocation_clean.csv", index=False)

print("✅ All cleaned CSVs saved!")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

q = """
SELECT DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
       SUM(oi.price + COALESCE(oi.freight_value,0)) AS revenue
FROM ecommerce.order_items oi
JOIN ecommerce.orders o USING (order_id)
GROUP BY month
ORDER BY month;
"""
monthly = pd.read_sql(q, engine)

plt.figure(figsize=(12,6))
plt.plot(monthly["month"], monthly["revenue"], marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month"); plt.ylabel("Revenue")
plt.xticks(rotation=45); plt.tight_layout()
plt.show()
```

```python
q = """
SELECT o.customer_id,
       SUM(oi.price + COALESCE(oi.freight_value,0)) AS total_spent
FROM ecommerce.order_items oi
JOIN ecommerce.orders o USING (order_id)
GROUP BY o.customer_id
ORDER BY total_spent DESC
LIMIT 10;
"""
top_customers = pd.read_sql(q, engine)

plt.figure(figsize=(10,6))
plt.barh(top_customers["customer_id"][::-1], top_customers["total_spent"][::-1])
plt.title("Top 10 Customers by Spend")
plt.xlabel("Total Spent"); plt.ylabel("Customer ID")
plt.tight_layout(); plt.show()
```

```python
q = """
SELECT LOWER(payment_type) AS payment_type,
       COUNT(*) AS payments,
       SUM(payment_value) AS amount
FROM ecommerce.payments
GROUP BY LOWER(payment_type)
ORDER BY amount DESC;
"""
paymix = pd.read_sql(q, engine)

fig, ax = plt.subplots(1,2, figsize=(14,5))
ax[0].bar(paymix["payment_type"], paymix["payments"])
ax[0].set_title("Payments Count by Type"); ax[0].set_ylabel("Count")
ax[1].bar(paymix["payment_type"], paymix["amount"])
ax[1].set_title("Payment Value by Type"); ax[1].set_ylabel("Total Value")
for a in ax: a.set_xlabel("Payment Type"); a.tick_params(axis='x', rotation=30)
plt.tight_layout(); plt.show()
```

```python
q = """
SELECT DATE_TRUNC('month', order_purchase_timestamp) AS month,
       AVG(EXTRACT(EPOCH FROM (order_delivered_customer_date - order_purchase_timestamp)) / 86400.0) AS avg_days
FROM ecommerce.orders
WHERE order_delivered_customer_date IS NOT NULL
GROUP BY month
ORDER BY month;
"""
delivery = pd.read_sql(q, engine)

plt.figure(figsize=(12,6))
plt.plot(delivery["month"], delivery["avg_days"], marker="o")
plt.title("Average Delivery Time (Days) per Month")
plt.xlabel("Month"); plt.ylabel("Avg Days")
plt.xticks(rotation=45); plt.tight_layout()
plt.show()
```
